# CANADA

```python
#@title SCRAPING OF CANADIAN PRESIDENT SPEECHS

def extract_statement ():

  dataframe = []
  driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
  all_data = []
  for page in range(0, 40):
    pagina ='https://pm.gc.ca/en/news/statements?page=' + str(page)
    print(f'Download the page {pagina}')
    driver.get(pagina)  
    sleep(randint(5,15))
    soup = bts(driver.page_source, 'html.parser')
    item_list = driver.find_elements_by_class_name('news-teaser')
    table = []

    #extract data from pages
    for i in range(len(item_list)):
      item = item_list[i]
      url = item.find_element_by_tag_name("a").get_property("href")
      date = item.find_element_by_class_name('location-date').text
      title = item.find_element_by_tag_name("a").text
      r = {
           'url': url,
           'date': date,
           'title': title
          }
      #Append the link to list links
      table.append(r)
    all_data.extend(table)
  data = pd.DataFrame(all_data)
  data.to_excel('/content/drive/MyDrive/Colab Notebooks/anahi_project/canada_data_information.xlsx', index=False)
  driver.close()
  return data
df_canada = extract_statement()
```

```python
#@title ZIP ALL FILES
import os
import zipfile
os.chdir('/content/drive/MyDrive/Colab Notebooks/anahi_project')
    
def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file), 
                       os.path.relpath(os.path.join(root, file), 
                                       os.path.join(path, '..')))
      
zipf = zipfile.ZipFile('australia.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('australia/', zipf)
zipf.close()
```
